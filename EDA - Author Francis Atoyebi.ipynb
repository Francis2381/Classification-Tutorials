Exploratory Data Analysis or EDA is basically as the name depicts the process of looking at or exploring your data with little or no modification to the data itself. EDA is the first step in a data analysis process and also a very important step that is simply because:
•	It gives you an idea of what kind of data you are working with that is data type and structure
•	It tells a story about your data whether numerical or not, tells if you have missing data or not, the number of missing data, mean, median and so many other summary statistics
•	It also guides you on what step you should take when ready to do data cleaning, manipulation (or feature engineering) or the kind of analysis you should carry out.

```python
#from the last 2 outputs above from both .head() and .info() function, we suspect that some of our data might be having incorrect data types so lets confirm that
def ucount(df, columns):
    import numpy as np
    for col in columns:
        num = df[col].unique()
        ife = num.tolist()
        types = len(ife)
        print(col +' has ' + str(types) + ' unique values ')
```

```python
columns=['symboling', 'normalized-losses', 'make-id', 'fuel-type', 'aspiration',
       'num-of-doors', 'body-style', 'drive-wheels', 'engine-location',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg', 'price']
ucount(df=autos, columns=columns)
```

```python
def catconverter(df, columns):
    '''convert columns of numeric or object type to categorical'''
    for col in columns:
        df[col] = df[col].astype('category')
    return df[columns].info()
```

```python
def describe(df, col): #Creating a function that takes in 2 arguments the dataframe name and column name in string to compute the summary statistics of that column
    # Compute the summary stats
    desc = df[col].describe()
    # Change the name of the 50% index to median
    indx = desc.index.tolist() #converting the index of describe function to a list
    indx[5] = 'median' #changing the value at index 5 from 50% to median
    desc.index = indx #assigning the updated index as the new index of describe whenever this function is used
    return desc

#let's test our new function
describe(autos, 'price') #you can play with this using other columns
```

#### Box Plot

```python
#Plotting a single box plot can be done this way
autos.boxplot(column='price', by='fuel-type')
```

```python
## Create boxplots of data
def auto_boxplot(df, plot_cols, by):
    %matplotlib inline
    import matplotlib.pyplot as plt
    for col in plot_cols:
        fig=plt.figure(figsize=(8,6))
        ax = fig.gca()
        df.boxplot(column = col, by = by, ax = ax)
        ax.set_title('Box plots of ' + col + ' by ' + by)
        ax.set_ylabel(col)
    return by
```

```python
colsss = ['symboling', 'normalized-losses', 'make-id', 'wheel-base', 'length', 'price']
auto_boxplot(autos, colsss, "fuel-type")
```
